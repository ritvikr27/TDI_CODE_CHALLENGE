```python
avg_units_651= sum_651/count_651
print('Average units called for 651: ', avg_units_651)
avg_units_111= sum_111/count_111
print('Average units called for 111: ', avg_units_111)

#Ratio of avg num of units called for a CODE 111 to CODE 651
ratio_of_avg= avg_units_111/avg_units_651
print('Ratio of avg num of units called for a CODE 111 to CODE 651: ', ratio_of_avg)
```

```python
#How many times more likely is an incident in Staten Island a false call compared to in Manhattan?
#Creating a new dataframe for giving a clear view of the data being used
df3=pd.DataFrame(data, columns= (['BOROUGH_DESC']))
df3=pd.concat([df1,df3], axis=1)
print(df3.head(20))
```

```python
#False Call Rate for Staten Island
count_staten= len(df3[(df3.BOROUGH_DESC== '3 - Staten Island')])
print('Total calls from Staten Island: ', count_staten)

false_count_staten= len(df3[(df3.BOROUGH_DESC =='3 - Staten Island') & (df3.INCIDENT_CODE == '710')])
print('False calls from Staten Island: ', false_count_staten)

false_rt_staten= false_count_staten/count_staten
print('False call rate from Staten Island: ', false_rt_staten)


#False Call Rate for Manhattan
count_manhattan= len(df3[(df3.BOROUGH_DESC== '1 - Manhattan')])
print('Total calls from Manhattan: ', count_manhattan)

false_count_manhattan= len(df3[(df3.BOROUGH_DESC =='1 - Manhattan') & (df3.INCIDENT_CODE == '710')])
print('False calls from Manhattan: ', false_count_manhattan)

false_rt_manhattan= false_count_manhattan/count_manhattan
print('False call rate from Manhattan: ', false_rt_manhattan)
```

```python
#Ratio of false call rates from Staten Island to Manhattan
false_rate= false_rt_staten/false_rt_manhattan
print('Ratio of False Call Rates of Staten Island to that of Manhattan: ', false_rate)
```
